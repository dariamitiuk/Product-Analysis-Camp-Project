```python
all_products = ifood_df.filter(like="Mnt")
all_products
```

```python
all_products
```

```python
popular_products_by_number = pd.DataFrame({'Amount_of_buying' : [all_products.query('MntWines > 0').count()[0],
                                all_products.query('MntFruits > 0').count()[0],
                                all_products.query('MntMeatProducts > 0').count()[0],
                                all_products.query('MntFishProducts > 0').count()[0],
                                all_products.query('MntSweetProducts > 0').count()[0],
                                all_products.query('MntGoldProds > 0').count()[0]]})
popular_products_by_number
```

```python
plt.title("Goods purchase number")
sns.barplot(x=['Wines', 'Fruits', 'Meat', 'Fish', 'Sweets', 'Gold'], 
            y=popular_products_by_number['Amount_of_buying'])
plt.ylabel("Number of purchases")
```

```python
popular_products_by_percent = pd.DataFrame({'Percent_in_receipt' : 
                                            [all_products['MntWines'].mean() / all_products['MntTotal'].mean() * 100,
                                            all_products['MntFruits'].mean() / all_products['MntTotal'].mean() * 100,
                                            all_products['MntMeatProducts'].mean() / all_products['MntTotal'].mean() * 100,
                                            all_products['MntFishProducts'].mean() / all_products['MntTotal'].mean() * 100,
                                            all_products['MntSweetProducts'].mean() / all_products['MntTotal'].mean() * 100,
                                            all_products['MntGoldProds'].mean() / all_products['MntTotal'].mean() * 100]})
popular_products_by_percent
```

```python
all_products['PercentWines'] = all_products['MntWines'] / all_products['MntTotal'] * 100
all_products['PercentFruits'] = all_products['MntFruits'] / all_products['MntTotal'] * 100
all_products['PercentMeat'] = all_products['MntMeatProducts'] / all_products['MntTotal'] * 100
all_products['PercentFish'] = all_products['MntFishProducts'] / all_products['MntTotal'] * 100
all_products['PercentSweets'] = all_products['MntSweetProducts'] / all_products['MntTotal'] * 100
all_products['PercentGold'] = all_products['MntGoldProds'] / all_products['MntTotal'] * 100
```

```python
all_products.head()
```

```python
popular_products_by_percent_2 = pd.DataFrame({'Percent_in_receipt' : 
                                            [all_products['PercentWines'].mean(),
                                            all_products['PercentFruits'].mean(),
                                            all_products['PercentMeat'].mean(),
                                            all_products['PercentFish'].mean(),
                                            all_products['PercentSweets'].mean(),
                                            all_products['PercentGold'].mean()]})
popular_products_by_percent_2
```

```python
plt.title("Percentage of products in receipt")
sns.barplot(x=['Wines', 'Fruits', 'Meat', 'Fish', 'Sweets', 'Gold'], 
            y=popular_products_by_percent['Percent_in_receipt'])
plt.ylabel("Percent")
```

```python
plt.title("Percentage of products in receipt")
sns.barplot(x=['Wines', 'Fruits', 'Meat', 'Fish', 'Sweets', 'Gold'], 
            y=popular_products_by_percent_2['Percent_in_receipt'])
plt.ylabel("Percent")
```

To find out the most popular products, i used 2 parameters:

i) how often the good gets in the receipt

ii) product price : total ratio in percent
I did it in 2 ways to compare the results. 
First: to divide product mean value of all customers by mean value total.
Second: to count product percentage of each customer and then find mean percent.
The results are noticeably different only for gold.

So, we can see that the most popular products in our dataset are wine, meat and gold products.

Middle year people visit this shop the most often, but average sum of receipt is about the same for every year group.

Number of total purchases is influenced by the income level of the customers.

As visiting web site in 80% ends in a purchase, more deals should be promoted in web to target customers' web purchases.

Wine, meat and gold products people buy the most often and pay for these the biggest sum.
